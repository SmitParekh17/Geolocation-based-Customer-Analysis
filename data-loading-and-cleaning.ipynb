```python
sum_list = ['Labour Total','OSL Total','Parts Total','Recovrbl Exp','Total Amt Wtd Tax.']

cols = ['Area / Locality', 'CITY', 'Cust Type', 'Customer No.', 'District', 
        'Gate Pass Time', 'Invoice Date', 'Invoice No', 'Invoice Time','JobCard Date', 
        'JobCard Time', 'KMs Reading', 'Labour Total', 'Make', 'Misc Total', 'Model',
        'OSL Total', 'Order Type', 'Parts Total', 'Pin code', 'Plant', 'Plant Name1', 
        'Print Status', 'Recovrbl Exp', 'Regn No', 'Technician Name', 'Total Amt Wtd Tax.', 'User ID']

value = []
for i in cols:
  if i not in sum_list:
    value.append('first')
  else:
    value.append('sum')
print(value)
dict_agg=dict(zip(list(cols),value))
dict_agg
```

```python
invoice_data_copy = invoice_data.copy()
```

```python
invoice_data = invoice_data.groupby(by = 'Job Card No').agg(dict_agg)
print(invoice_data.shape)
```
